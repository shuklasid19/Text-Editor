```python
#importing the library
#imports the  files dialog for saving
# * imports everything
import datetime
from tkinter import filedialog
from tkinter import *
#gives a dialog 
#forground and background color
#asks yes and no
from tkinter.messagebox import askyesno
from tkinter.colorchooser import askcolor
import webbrowser


#initializing
root = Tk()

#all the methods to perform operations
#we select all text and we save and write in that file
def New_File():
    if askyesno('Notepad',"Save Existing work?" ):
        filename = filedialog.asksaveasfilename()
        if filename:
            alltext = text.get(1.0 , END) #saves frm 1st line
            open(filename, 'w').write(alltext)
    
    #open file in reading mode and ask if u want ot open existing file
    if askyesno("Notepad", "open Existing File?"):
        #delete the file
        text.delete(1.0 , END)
        file = open(filedialog.askopenfilename(), "r")
        #if file name is not empty
        if file != "":
            txt = file.read()
            text.insert(INSERT, txt)
        else:
            pass
    else:
        text.delete(1.0, END)
 

#starts with 1st line for reading the doc
def Open_File():
    text.delete(1.0, END)
    file = open(filedialog.askopenfilename(), "r")
    if file != "":
        txt = file.read()
        text.insert(INSERT, txt)
        
#save the text 
def Save_As():
    filename = filedialog.asksaveasfilename()
    if filename:
        alltext = text.get(1.0, END)
        open(filename, 'w').write(alltext)
        
#close the notepad
#and asks for save
def Close():
    if askyesno('Notepad',"Save Existing work?" ):
        filename = filedialog.asksaveasfilename()
        if filename:
            alltext = text.get(2.0 , END)
            open(filename, 'w').write(alltext)
        root.destroy()
    else:
        root.destroy()
    
    

#cut the text and copies into clipboard
def Cut():
    text.clipboard_clear()
    text.clipboard_append(text.selection_get)
    sel = text.get(SEL_FIRST, SEL_FIRST)
    text.delete(SEL_FIRST, SEL_LAST)

#copies the text to the clipboard
def Copy():
    text.clipboard_clear()
    text.clipboard_append(text.selection_get())

#for pasting 
def Paste():
    try:
        txt = text.selection_get(selection = "CLIPBOARD")
        text.insert(INSERT, txt)
    except:
        pass

#erases the text
def Erase():
    sel = text.get(SEL_FIRST, SEL_LAST)
    text.delete(SEL_FIRST, SEL_LAST)

#clears the screen from 1st to end    
def Clear_Screen():
    text.delete(1.0, END)

#shows the date and time
def Date():
    data = datetime.date.today()
    text.insert(INSERT, data)

#can change the text color
def Text_color():
    (triple, color) = askcolor()
    
    if color:
        text.config(foreground=color)

#can chhange the format
def No_Format():
    text.config(font= ("Arial", 10))
    
#can change into bold
def Bold():
    current_tags = text.tag_names("sel.first")
    
    if "bt" in current_tags:
        text.tag_remove("bt", "sel.first", "sel.last")
        text.tag_config("bt", font = ("Arial", 10, "bold"))
    else:
        text.tag_add("bt", "sel.first", "sel.last")
        text.tag_config("bt", font=("Arial", 10, "bold"))

#change text into italic
def Italic():
    current_tags = text.tag_names("sel.first")
    
    if "bt" in current_tags:
        text.tag_remove("bt", "sel.first", "sel.last")
        text.tag_config("bt", font = ("Arial", 10, "italic"))
    else:
        text.tag_add("bt", "sel.first", "sel.last")
        text.tag_config("bt", font=("Arial", 10, "italic"))
           

#it creates text covered in black
def Underline():
    text.tag_add("here" ,"1.0" , "1.4")
    #text.tag_add("bte" , "1.5", "1.8")
    #text.tag_config("bte", font = ("Arial", 10, "underline"))
    text.tag_config("here", background ="black", foreground="White")

    
#can change the background    
def Background():
    (triple, color) = askcolor()
    if color:
        text.config(background = color)

def User_Help():
    pass

def Online_Help():
    webbrowser.open("https://www.google.com/")

#for submenu's weve cereated command
#title of the notepad
if __name__ == "__main__":
    root.title('Notepad')
    main_menu = Menu(root)
    submenu = Menu(root, tearoff=0)
    commands = Menu(root)
    root.config(menu = main_menu)
    
    #were creating objects(menu class) for our cascade window==root
    main_menu.add_cascade(label= "File", menu=commands)
    commands.add_command(label = "New_File", command = New_File)
    commands.add_command(label = "Open...", command=Open_File )
    commands.add_command(label = "Save_As...", command = Save_As)
    commands.add_command(label = "Close" , command=Close)
    
    
    #object for the edit menu
    edit_menu = Menu(root)
    main_menu.add_cascade(label = "Edit", menu=edit_menu)
    edit_menu.add_command(label = "cut", command=Cut)
    edit_menu.add_command(label = "Copy", command = Copy)
    edit_menu.add_command(label = "Paste", command=Paste)
    edit_menu.add_separator()
    edit_menu.add_command(label ="Delete", command = Erase)
    edit_menu.add_command(label = "Clear Screen" ,command = Clear_Screen)
    
    #another object which shows the date 
    insert_in_menu = Menu(root)
    main_menu.add_cascade(label = "Insert", menu = insert_in_menu)
    insert_in_menu.add_command(label = "Current Date", command=Date)
    
    #another obj changes the format
    change_format = Menu(root)
    main_menu.add_cascade(label = "Format", menu=change_format)
    change_format.add_command(label = "Font", command = Text_color )
    change_format.add_separator()
    change_format.add_command(label = "No Format", command=No_Format)
    change_format.add_command(label = "Bold", command = Bold)
    change_format.add_command(label = "Italic", command = Italic)
    change_format.add_command(label = "Underline", command=Underline )
    
    #u can change the text
    Personalize = Menu(root)
    main_menu.add_cascade(label = "Personalize" , menu = Personalize)
    Personalize.add_command(label = "Background", command=Background)
    
    #opens the browser
    User_Help = Menu(root)
    main_menu.add_cascade(label = "help", menu=User_Help)
    User_Help.add_cascade(label = "Online Help ", command=Online_Help)
    
    
    
    #scollbar in y-axis
    text = Text(root, height=40, width=100, font=("Arial", 10))
    scrollbar = Scrollbar(root, command=text.yview)
    scrollbar.config(command = text.yview)
    text.config(yscrollcommand = scrollbar.set)
    scrollbar.pack(side = RIGHT, fill=Y)
    text.pack()
    root.resizable(0, 0)
    root.mainloop()
    
    
#finalizing
root.mainloop()
```
